```python
import numpy as np
import matplotlib.pyplot as plt

alpha = 1
c0 = 1
t = np.arange(0, 1.1, 0.1)
f = c0 * np.exp(alpha * t)
```

```python
plt.plot(t, f)
plt.xlabel('t')
plt.ylabel('f')
plt.title('Exponential Growth')
```

```python
plt.show()
```

```python
lmbda = 1
c0 = 10
t = np.arange(0, 1.1, 0.1)
f = c0 * np.exp(-lmbda * t)

```

```python
plt.plot(t, f)
plt.xlabel('t')
plt.ylabel('f')
plt.title('Exponential Decay')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

c0 = 1
lambdas = [1, 2, 3]  # List of lambda values

t = np.arange(0, 1.1, 0.1)

plt.figure()  # Create a new figure

for lmbda in lambdas:
    f = c0 * np.exp(-lmbda * t)
    plt.plot(t, f, label=f'lambda = {lmbda}')

plt.xlabel('t')
plt.ylabel('f')
plt.title('Exponential Decay with Different Lambdas')
plt.legend()
plt.grid(True)
plt.show()
```

**Ifs and Loops in Python**

```python
import numpy as np
import matplotlib.pyplot as plt

c0 = 1
t = np.arange(0, 1.1, 0.1)
lambda_values = range(1, 6)  # Lambda values from 1 to 5

plt.figure()  # Create a new figure

for lmbda in lambda_values:
    f = c0 * np.exp(-lmbda * t)
    plt.plot(t, f, label=str(lmbda))

plt.xlabel('t')
plt.ylabel('f')
plt.title('Exponential Decay with Different Lambdas')
plt.legend()
plt.grid(True)
plt.show()
```
